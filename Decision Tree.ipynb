# Implementing Entropy Function

```python
import math
def entropy(S):
    #should find probablities from data set S
    target_values=S[S.columns[-1]].value_counts()
    tot_tuples=len(S.index)
    prob=[i/tot_tuples for i in target_values]
    sum=0
    for pi in prob:
        sum += -pi*math.log2(pi)
    return sum
```

```python
#Testing entropy function
print(entropy(data))#output should be 0.98
```

Now, Implementing Information Gain function

```python
#Information gain
def IG(S,A):
    categ_values=S[A].unique()
    mod_S=len(S.index)
    sum=0
    for i in categ_values:
        sum+=(len(S[S[A]==i].index)/mod_S)*entropy(S[S[A]==i])
    return entropy(S)-sum
    
```

# Decision Tree

```python
import pandas as pd
import math

class Node:
    def __init__(self,data,child):
        self.child=child
        self.data=data

class DecisionTreeClass:
    def entropy(self,S):
        #should find probablities from data set S
        target_values=S[S.columns[-1]].value_counts()
        tot_tuples=len(S.index)
        prob=[i/tot_tuples for i in target_values]
        sum=0
        for pi in prob:
            sum += -pi*math.log2(pi)
        return sum

    def IG(self,S,A):
        categ_values=S[A].unique()
        mod_S=len(S.index)
        sum=0
        for i in categ_values:
            sum+=(len(S[S[A]==i].index)/mod_S)*self.entropy(S[S[A]==i])
        return self.entropy(S)-sum

    def generate_decision_tree(self,D,attr_list):
        is_same=len(D[D.columns[-1]].unique())
        if is_same==1:
            return Node(D[D.columns[-1]].unique()[0],[])
        if(len(attr_list)==0):
            return Node("majority class",[])
        max=-100
        for attr in attr_list:
            tmp=self.IG(D,attr)
            if(tmp>max):
                max=tmp
                select_attr=attr
        attr_list.remove(select_attr)
        N=Node(select_attr,[])
        splitting_criterion=D[select_attr].unique()
        for j in splitting_criterion:
            D_J=D[D[select_attr]==j]
            if len(D_J.index)==0:
                return Node("Majority class",[])
            else:
#                 print('Entered')
                N.child.append(self.generate_decision_tree(D_J,attr_list))
        return N
        
    def classifydata(self,data):
        attr_ls=data.columns.tolist()
        attr_ls.remove('buys_computer')
        k=self.generate_decision_tree(data,attr_ls)
        return k
    
    def traverse(self,node, path = []):
        path.append(node.data)
        if len(node.child) == 0:
            print(path)
            path.pop()
        else:
            for child in node.child:
                self.traverse(child, path)
            path.pop()
```
